```python
import pandas as pd
from sklearn.model_selection import train_test_split
from rake_nltk import Rake
```

```python
df = pd.read_csv("follow_ML_data.csv")
df.head()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
df.drop("valid followUp (y/n)", axis = 1, inplace = True)
df.drop("rank", axis = 1, inplace = True)
df.drop("Unnamed: 6", axis = 1, inplace = True)
df.drop("question", axis = 1, inplace = True)
```

```python
X = df.copy()
X.drop("generated_fq", axis = 1, inplace = True)
y = df
y.drop("answer", axis = 1, inplace = True)
X.head()
```

```python
r = Rake()
X['keywords'] = 'default'
y['keywords'] = 'default'
X.shape
```

```python
count = 0
for i in X.itertuples():
    text = i[1]
    r.extract_keywords_from_text(text)
    l1 = r.get_ranked_phrases()
    phrases = ""
    for j in l1:
        phrases = phrases + j + " "
    X.loc[count, 'keywords'] = phrases
    count += 1
    
count = 0
for i in y.itertuples():
    text = i[1]
    r.extract_keywords_from_text(text)
    l1 = r.get_ranked_phrases()
    phrases = ""
    for j in l1:
        phrases = phrases + j + ", "
    phrases.strip(', ')
    y.loc[count, 'keywords'] = phrases
    count += 1
    
X.head()
```

```python
y.head()
```

```python
xx = X.drop('answer', axis = 1, inplace = False)
y.drop('keywords', axis = 1, inplace = True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(xx, y, test_size=0.2, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=1)
Test, garbage = train_test_split(X, test_size = 0.2, random_state = 1)
garbage.head()
```

```python
X_train.to_csv(r'src-train.txt', header=False, index = False, sep='\n', mode='w')
X_val.to_csv(r'src-val.txt', header=False, index = False, sep='\n', mode='w')
X_test.to_csv(r'src-test.txt', header=False, index = False, sep='\n', mode='w')
y_train.to_csv(r'tgt-train.txt', header=False, index = False, sep='\n', mode='w')
y_val.to_csv(r'tgt-val.txt', header=False, index = False, sep='\n', mode='w')
y_test.to_csv(r'tgt-test.txt', header=False, index = False, sep='\n', mode='w')
```
