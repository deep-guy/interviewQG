```python
from rake_nltk import Rake
import pandas as pd
```

```python
df = pd.read_csv("followML_extra_ratings.csv")
df.head()
```

```python
df.drop("Grammar", axis = 1, inplace = True)
df.drop("Relevance", axis = 1, inplace = True)
df.drop("followUp", axis = 1, inplace = True)
df.head()
```

```python
df1 = df["question"]
df2 = df["answer"]
df1.head()
```

```python
df1.to_csv(r'questions.txt', header=False, index=None, sep=' ', mode='a')
df2.to_csv(r'answers.txt', header=False, index=None, sep=' ', mode='a')
```

```python
r = Rake()
mydict = {}
df['question_keywords'] = "default"
df['answer_keywords'] = 'default'
df.head()
    
```

```python
count = 0
for i in df.itertuples():
    text = i[1]
    r.extract_keywords_from_text(text)
    l1 = r.get_ranked_phrases()
    phrases = ""
    for j in l1:
        phrases = phrases + j + ", "
    phrases.strip(', ')
    df.loc[count, 'question_keywords'] = phrases
    count += 1
    
count = 0
for i in df.itertuples():
    text = i[2]
    r.extract_keywords_from_text(text)
    l1 = r.get_ranked_phrases()
    phrases = ""
    for j in l1:
        phrases = phrases + j + ", "
    phrases.strip(', ')
    df.loc[count, 'answer_keywords'] = phrases
    count += 1
    
df.head()
    
```

```python
df.to_csv(r'extracted_keywords.csv', header=True, sep=',', mode='a')
```
